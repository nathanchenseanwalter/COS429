```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage
```

```python
img = cv2.imread('Lenna_(test_image).png',1)
```

2

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])
plt.show()
```

3

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

4

```python
height, width = gray.shape
print(height, width)
```

5

```python
height_center = height//2
width_center = width//2
cropped = gray[height_center-50:height_center+50, width_center-50:width_center+50]
plt.imshow(cropped, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

6

```python
altered = np.copy(cropped)
for i in range(len(altered)):
    for j in range(len(altered[i])):
        if (i % 20 == 0) & (j % 10 == 0) & (i > 0) & (j > 0):
            altered[i][j] = 0
            
plt.imshow(altered, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

7

```python
combined = np.zeros((100, 100*2), dtype=np.uint8)
combined[:, 100:200] = cv2.flip(altered, 0)
combined[:, 0:100] = altered
plt.imshow(combined, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

8

```python
cv2.imwrite('combined.png', combined)
```
